```python
import tensorflow as tf
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout,Input,Flatten,Dense,MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# tf.test.is_gpu_avaiilable()
physical_devices =tf.config.list_physical_devices('GPU')
print("Num GPUs:", len(physical_devices))
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,rotation_range=0.2,
                              shear_range=0.2,
                              zoom_range=0.2,width_shift_range=0.2,height_shift_range=0.2,validation_split=0.2)
```

```python
train_data=train_datagen.flow_from_directory(r'C:\Users\91706\Desktop\druken driver detection project\prepare data\train data',target_size=(80,80),batch_size=8,class_mode='categorical',subset='training')
validation_data=train_datagen.flow_from_directory(r'C:\Users\91706\Desktop\druken driver detection project\prepare data\train data',target_size=(80,80),batch_size=8,class_mode='categorical',
                                                 subset='validation')
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
test_data=test_datagen.flow_from_directory(r'C:\Users\91706\Desktop\druken driver detection project\prepare data\test data',target_size=(80,80),batch_size=8,class_mode='categorical')
```

```python
base_model= InceptionV3(include_top=False,weights='imagenet',input_tensor=Input(shape=(80,80,3)))
head_model=base_model.output
```

```python
base_model.summary()
```

```python

head_model=Flatten()(head_model)
head_model=Dense(64,activation='relu')(head_model)
head_model=Dropout(0.5)(head_model)
head_model=Dense(2,activation='softmax')(head_model)

model=Model(inputs=base_model.input,outputs=head_model)

for layer in base_model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping,ReduceLROnPlateau
```

```python
checkpoint=ModelCheckpoint(r'C:\Users\91706\Desktop\druken driver detection project\model_checkpoint\model.h5',monitor='val_loss',save_best_only=True,verbose=3)

earlystop=EarlyStopping(monitor='val_loss',patience=7,verbose=3,restore_best_weights=True)
learning_rate=ReduceLROnPlateau(monitor='val_loss',patience=3,verbose=3)
callbacks=[checkpoint,earlystop,learning_rate]
```

```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
import keras
from PIL import Image
from sklearn.preprocessing import LabelEncoder
```

```python
batchsize=8
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit_generator(train_data,steps_per_epoch=train_data.samples//batchsize,
                   validation_data=validation_data,
                   validation_steps=validation_data.samples//batchsize,
                   callbacks=callbacks,epochs=5)
```

```python
#model evaluation
```

```python
acc_tr,loss_tr=model.evaluate_generator(train_data)
print(acc_tr)
print(loss_tr)
```

```python
acc_vr,loss_vr=model.evaluate_generator(validation_data)
print(acc_vr)
print(loss_vr)
```

```python
acc_te,loss_te=model.evaluate_generator(test_data)
print(acc_te)
print(loss_te)
```
